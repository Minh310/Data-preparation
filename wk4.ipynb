```python
import pandas as pd 
```

Ex1

```python
# a.
dt_aapl = pd.read_csv('aapl.csv', index_col = [0], parse_dates = [0])
dt_amzn = pd.read_csv('amzn.csv', index_col = [0], parse_dates = [0])
dt_fb = pd.read_csv('fb.csv', index_col = [0], parse_dates = [0])
dt_goog = pd.read_csv('goog.csv', index_col = [0], parse_dates = [0])
dt_nflx = pd.read_csv('nflx.csv', index_col = [0], parse_dates = [0])
```

```python
# b.
dt_aapl['ticker'] = 'APPL'
dt_amzn['ticker'] = 'AMZ'
dt_fb['ticker'] =  'FB'
dt_goog['ticker'] = 'GOOG'
dt_nflx['ticker'] = 'NETF'
```

```python
# c.
faang = pd.concat([dt_aapl,dt_amzn,dt_fb,dt_goog,dt_nflx], axis = 0, join = 'inner').sort_index()
faang.head()
```

```python
# d.
faang.to_csv('faang.csv')
```

```python
# e.
faang.index = pd.to_datetime(faang.index)
faang['volume'] = faang['volume'].astype(int)
faang.head()
```

```python
faang.sort_values(by=['date', 'ticker'])
faang.head()
```
