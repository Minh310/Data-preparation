```python
dt = pd.read_csv("earthquakes.csv")
dt.head()
```

```python
bins = pd.IntervalIndex.from_tuples([(0, 1), (1, 2), (4, 7)])
pd.cut(dt[dt.magType=="ml"].mag, bins, labels=["first","second","third"]).value_counts()
```
